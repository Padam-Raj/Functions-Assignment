Ques 1. Write a python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def even_sum(nums):
    if type(nums)==list:
        add_even=0
        for i in nums:
            if i%2==0:
                add_even+=i
        return add_even
    else:
        return "kindly provide list of numbers"

even_sum([3,4,5,6,7,8,9,10])
```

Ques 2. Create a python function that accepts a string and returns the reverse of that string.

```python
def str_rev(st):
    if type(st)==str:
        return st[::-1]
    else:
        return "kindly provide a string to reverse"

str_rev("My name is Padam Raj")
```

Ques 3. Implement a python function that takes a list of integers and returns a new list containing the squares of each number.

```python
def sq_nums(nums):
    sq_list=[]
    for i in nums:
        sq_list.append(i**2)
    return sq_list


sq_nums([2,3,4,5,6,7,8,9])
```

Ques 4. Write a python function that checks if a given number is prime or not from 1 to 200.

```python
import math

def prime_num(n):
    
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

num= int(input("Please enter a number between 1 to 200: "))
if prime_num(num):
    print("It is a prime Number")
else:
    print("It is not a prime Number")
```

Ques 5. Create an iterator class in python that generates the fibonacci sequence up to a specified number of terms.

```python
def fib(n):
    a=0
    b=1
    fib_ser=[]
    for i in range(n):
        fib_ser.append(a)
        a, b= b, a+b
    return fib_ser


fib(10)
```

Ques 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def file_read(path):
    with open(path, 'r', encoding='utf-8') as f:
        for l in f:
            yield l.rstrip('\n')

s=file_read("demo.txt")
next(s)
```

Ques 10. Create a python program that uses 'filter()' to remove all the vowels from a given string.

```python
def remove_vowels(s):
    return ''.join(filter(lambda ch: ch.lower() not in ['a', 'e', 'i', 'o', 'u'],s))

Str_user= input("Please enter a string- ")
print(f"String without vowels-{remove_vowels(Str_user)}")
```
